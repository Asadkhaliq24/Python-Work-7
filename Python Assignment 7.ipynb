```python
class Employee:
    def __init__(self, name, emp_id, salary, department):
        self.name = name
        self.id = emp_id
        self.salary = salary
        self.department = department

    def calculate_salary(self, salary, hours_worked):
        overtime = 0
        if hours_worked > 50:
            overtime = hours_worked - 50
        self.salary = self.salary + (overtime * (self.salary / 50))

    def assign_department(self, emp_department):
        self.department = emp_department

    def print_employee_details(self):
        print("\nName: ", self.name)
        print("ID: ", self.id)
        print("Salary: ", self.salary)
        print("Department: ", self.department)
        print("----------------------")


employee1 = Employee("ASAD", "E7876", 50000, "DATA ANALYST")
employee2 = Employee("SAAD", "E7499", 45000, "RESEARCH")
employee3 = Employee("WAQAS", "E7900", 50000, "SALES")
employee4 = Employee("SHARIQ", "E7698", 55000, "OPERATIONS")

print("Original Employee Details:")
employee1.print_employee_details()
employee2.print_employee_details()
employee3.print_employee_details()
employee4.print_employee_details()

# Change the departments of employee1 and employee4
employee1.assign_department("OPERATIONS")
employee4.assign_department("SALES")

# Now calculate the overtime of the employees who are eligible:
employee2.calculate_salary(45000, 52)
employee4.calculate_salary(45000, 60)

print("Updated Employee Details:")
employee1.print_employee_details()
employee2.print_employee_details()
employee3.print_employee_details()
employee4.print_employee_details()
```

***Make a method called describe_employee() that prints the detail Information about employee, a method update_ year_of_experience(), and a method called increase_salary() that do increamnet in it's salary***

```python
class Employee:
    def __init__(self, name, position, salary, years_of_experience):
        self.name = name
        self.position = position
        self.salary = salary
        self.years_of_experience = years_of_experience

    def describe_employee(self):
        print(f"Employee Name: {self.name}")
        print(f"Position: {self.position}")
        print(f"Salary: ${self.salary:,.2f}")
        print(f"Years of Experience: {self.years_of_experience}")

    def update_year_of_experience(self, new_years_of_experience):
        if new_years_of_experience >= 0:
            self.years_of_experience = new_years_of_experience
        else:
            print("Years of experience cannot be negative.")

    def increase_salary(self, increment_amount):
        if increment_amount > 0:
            self.salary += increment_amount
        else:
            print("Increment amount must be positive.")

# Example usage:
employee = Employee("Asad Khaliq", "Data Analyst", 85000, 5)

# Print employee details
employee.describe_employee()

# Update years of experience
employee.update_year_of_experience(6)

# Increase salary
employee.increase_salary(5000)

# Print updated employee details
employee.describe_employee()
```

```python
class Employee:
    def __init__(self, name, salary, years_of_experience):
        self.name = name
        self.salary = salary
        self.years_of_experience = years_of_experience

    def describe_employee(self):
        return f"Name: {self.name}, Salary: ${self.salary}, Years of Experience: {self.years_of_experience}"

    def increase_salary(self, amount):
        self.salary += amount
        return self.salary

    def increase_years_of_experience(self, years):
        self.years_of_experience += years
        return self.years_of_experience

# Create an instance called employee101
employee101 = Employee("Alice", 50000, 5)

# Print the four attributes individually
print(f"Name: {employee101.name}")
print(f"Salary: ${employee101.salary}")
print(f"Years of Experience: {employee101.years_of_experience}")

# Call all its methods
print(employee101.describe_employee())

# Increase salary and years of experience
print(f"New Salary after increase: ${employee101.increase_salary(5000)}")
print(f"New Years of Experience after increase: {employee101.increase_years_of_experience(1)}")

# Create 10 different instances of Employee
employees = [
    Employee("Asad", 55000, 4),
    Employee("Ali", 60000, 6),
    Employee("Ahmer", 65000, 8),
    Employee("Asher", 70000, 2),
    Employee("Talha", 75000, 7),
    Employee("Saad", 80000, 10),
    Employee("Waqas", 85000, 3),
    Employee("Nasim Farhan", 90000, 5),
    Employee("Shariq Aziz", 95000, 9),
    Employee("Abdullah", 100000, 12)
]

# Call describe_employee() for each instance
for emp in employees:
    print(emp.describe_employee())

# Perform some updates with its methods
print("Updating Employees...")

for emp in employees:
    # Increase salary by $2000 and years of experience by 2
    print(f"Updated Salary for {emp.name}: ${emp.increase_salary(2000)}")
    print(f"Updated Years of Experience for {emp.name}: {emp.increase_years_of_experience(2)}")
```
